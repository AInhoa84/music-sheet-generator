## 1. Feature exploration

These files are a sequence of notes from C#3 to B5 played by different pianos

```python
y, sr = librosa.load("../Samples/Piano/PN_ab-ba_piano.mp3", mono=False)
```

```python
y = librosa.core.to_mono(y)
y = librosa.resample(y, sr, 8000)
y = y[np.argwhere(y >= 0.001)[0][0]:]
```

```python
plt.figure(figsize=(18,6))
plt.plot(y[:40000])
```

Each note takes up 16000 samples

```python
plt.figure(figsize=(18,6))
plt.plot(y[:16000])
```

```python
plt.figure(figsize=(18,6))
plt.plot(y[16000:16000*2])
```

```python
C_3 = y[:16000]
```

### 1.1. Zero crossing rate

```python
librosa.feature.zero_crossing_rate(C_3).mean()
```

```python
note_table = pd.read_csv("../Utilities/Note_table.tsv", header=0, sep="\t")
note_table[note_table["Note"] == "C#3"]
```

```python
note_table[note_table["Note"] == "B5"]
```

```python
notes = note_table.iloc[37:72]["Note"]
```

```python
zcr = []
for i in range(0, y.shape[0], 16000):
    zcr.append(librosa.feature.zero_crossing_rate(y[i:(i+16000)]).mean())
zcr = zcr[:-1] # The last chunk is just a tail
```

ZCR seems to increase with note frequency

```python
plt.figure(figsize=(18,6))
plt.plot(notes, zcr, 'rx', markersize=10)
plt.xlabel("Note")
plt.ylabel("Zero crossing rate")
```

```python
zcr = []
for i in range(0, y.shape[0], 16000):
    zcr.append(librosa.feature.zero_crossing_rate(y[i:(i+5000)]).mean())
zcr = zcr[:-1] # The last chunk is just a tail
```

It's interesting to see that with a shorter note duration the pattern is still there

```python
plt.figure(figsize=(18,6))
plt.plot(notes, zcr, 'rx', markersize=10)
plt.xlabel("Note")
plt.ylabel("Zero crossing rate")
```

### 1.2. Peak rate

```python
plt.figure(figsize=(18,6))
plt.plot(y[:20], 'x', markersize=10)
plt.plot(y[:20])
```

```python
c = 0
for j in range(1, 19):
    if ((y[j]>y[j-1]) and (y[j]>y[j+1])) or ((y[j]<y[j-1]) and (y[j]<y[j+1])):
        c += 1
print("Number of peaks: ", c)
```

```python
pr = []
for i in range(0, 16000*35, 16000):
    c = 0
    for j in range(i+1,i+15999):
        if ((y[j]>y[j-1]) and (y[j]>y[j+1])) or ((y[j]<y[j-1]) and (y[j]<y[j+1])):
            c += 1
    pr.append(c/16000)
```

```python
plt.figure(figsize=(18,6))
plt.plot(notes, pr, 'rx', markersize=10)
```

```python
F5 = y[28*16000:29*16000]
plt.figure(figsize=(18,6))
plt.plot(F5)
```

```python
c = 0
for j in range(1, 15999):
    if ((F5[j]>F5[j-1]) and (F5[j]>F5[j+1])) or ((F5[j]<F5[j-1]) and (F5[j]<F5[j+1])):
        c += 1
print("Peak rate for F5: ", c/16000)
```

```python
G_5 = y[31*16000:32*16000]
plt.figure(figsize=(18,6))
plt.plot(G_5)
```

```python
pr = []
for i in range(0, 16000*35, 16000):
    c = 0
    for j in range(i+1,i+4000):
        if ((y[j]>y[j-1]) and (y[j]>y[j+1])) or ((y[j]<y[j-1]) and (y[j]<y[j+1])):
            c += 1
    pr.append(c/4000)
```

The peak rate does seem to be influenced by the number of samples we take to compute it

```python
plt.figure(figsize=(18,6))
plt.plot(notes, pr, 'rx', markersize=10)
```

```python
plt.figure(figsize=(10,8))
plt.plot(zcr, pr, 'rx', markersize=10)
```

```python
from scipy.stats.stats import pearsonr
pearsonr(zcr,pr)
```
