## NN

```python
data["Note"] = librosa.note_to_midi(data["Note"]) - 49
```

```python
def create_xy(df, target_column):
    return df.drop(target_column, axis=1), df[target_column]

def split_data(df, target_column):
    X_train, y_train = create_xy(df.sample(round(0.8*df.shape[0])), target_column)
    df = df.drop(X_train.index)
    X_val, y_val = create_xy(df.sample(round(0.5*df.shape[0])), target_column)
    df = df.drop(X_val.index)
    X_test, y_test = create_xy(df, target_column)
    return X_train, y_train, X_val, y_val, X_test, y_test
```

```python
X_train, y_train, X_val, y_val, X_test, y_test = split_data(data.drop(["Piano", "Pred_note"], axis=1), "Note")
```

```python
model = keras.Sequential([
    keras.layers.Dense(256, input_shape=(16000,)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(35, activation=tf.nn.softmax)
])
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=15, verbose=2)
```

```python
val_loss, val_acc = model.evaluate(X_val, y_val)
val_acc
```

```python
NN_predictions = np.argmax(model.predict(X_val), axis=1)
CQT_predictions = []

for row in tqdm(X_val[["x{}".format(i) for i in range(16000)]].values):
    cqt = librosa.core.cqt(row, 8000, n_bins=80)
    CQT_predictions.append(librosa.amplitude_to_db(cqt, ref=np.max).mean(axis=1).argmax()+24-49)
CQT_predictions = np.array(CQT_predictions)
    
index_misclass_NN = (NN_predictions != y_val).values
index_misclass_CQT = (CQT_predictions != y_val).values
```

```python
(y_val[index_misclass_NN].values == CQT_predictions[index_misclass_NN]).sum()/index_misclass_NN.sum() * 100
```

```python
print("{:>8} {:>8} {:>8}".format("NN", "CQT", "y_val"))
for i, j, k in zip(NN_predictions[index_misclass_NN], 
                   CQT_predictions[index_misclass_NN], y_val[index_misclass_NN]):
    print("{:>8} {:>8} {:>8}".format(i, j, k))
```
