Podemos calcular la envolvente

```python
env_pos = []
env_neg = []
n = 200

for i in range(0, len(y), n):
    env_pos += n * [np.max(y[i:(n+i)])]
    env_neg += n * [np.min(y[i:(n+i)])]
```

```python
plt.plot(y)
plt.plot(env_pos, 'r')
plt.plot(env_neg, 'r')
plt.tight_layout()
```

```python
mean_pos = y[y >= 0].mean()
mean_neg = abs(y[y < 0].mean())
norm_y = y.copy

for i in range(len(norm_y)):
    if norm_y[i] >= 0:
        norm_y[i] = norm_y[i] * (mean_pos/env_pos)
    else:
        norm_y[i] = norm_y[i] * (mean_neg/env_neg)
```
