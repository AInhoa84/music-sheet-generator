```python
import librosa
import librosa.display
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
y, sr = librosa.load("mp3 Notes/a3.mp3")
```

Sampling rate in Hz

```python
sr
```

Waveform

```python
plt.figure(figsize=(18,6))
plt.plot(y)
```

We can calculate the envelope

```python
env_pos = []
env_neg = []
n = 200

for i in range(0, len(y), n):
    env_pos += n * [np.max(y[i:(n+i)])]
    env_neg += n * [np.min(y[i:(n+i)])]
```

```python
plt.figure(figsize=(18,6))
plt.plot(y)
plt.plot(env_pos, 'r')
plt.plot(env_neg, 'r')
plt.tight_layout()
```

```python
plt.figure(figsize=(18,6))
plt.plot(y[6110:6800])
plt.plot(0.015*np.sin(2*np.pi*110.25/sr*np.arange(700)), 'r')
plt.tight_layout()
```

```python
sr/200
```

```python
mean_pos = y[y >= 0].mean()
mean_neg = y[y < 0].mean()
norm_y = y.copy()

for i in range(len(norm_y)):
    if norm_y[i] >= 0:
        norm_y[i] = norm_y[i] * (mean_pos/env_pos[i])
    else:
        norm_y[i] = norm_y[i] * abs(mean_neg/env_neg[i])
```

```python
plt.figure(figsize=(18,6))
plt.plot(y[7000:8000])
plt.plot(norm_y[7000:8000], 'r', alpha=.6)
```

```python
plt.figure(figsize=(18,6))
plt.plot(norm_y[800:1600])
```

```python
plt.figure(figsize=(18,6))
plt.plot(norm_y[950:1200])
```

```python
S = librosa.feature.melspectrogram(y)
librosa.display.specshow(librosa.power_to_db(S, ref=np.max), y_axis="mel", x_axis="time")
plt.tight_layout()
```

```python
w = np.fft.fft(y)
freqs = np.fft.fftfreq(len(y))
plt.figure(figsize=(18,6))
plt.plot(sr*freqs[0:500],abs(w[0:500]))
```
